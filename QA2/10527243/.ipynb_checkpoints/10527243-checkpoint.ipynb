### kNN search
 data| Q | A | B | C | D | E 
-----|---|---|---|---|---|---     
R    |160|5  |10 |25 |100|125  
R2   |25 |50 |40 |10 |20 |60   

                             If QF = 40, what are the top-3 objects close to Q?

```python
# 點
class Node():
    def setNode(self, name, r1, r2):
        self.name = name
        self.r1 = r1
        self.r2 = r2
    
    def Inf(self):
        print("name : " + self.name)
        print("R1 : " + str(self.r1))
        print("R2 : " + str(self.r2))
        
    def nodeName(self):
        return self.name
    name = ''
    r1 = 0
    r2 = 0

# 每個點之間的結果
class Result():
    sideName = ''
    leftSideLen = 0
    rightSideLen = 0
    def setSideName_LeftLen_RightLen(self, n1, n2):
        self.sideName = n1.nodeName() + n2.nodeName() # 邊的名字
        
        # max fo 相減結果
        r1Result = abs(n1.r1 - n2.r1)
        r2Result = abs(n1.r2 - n2.r2)
        self.leftSideLen = max(r1Result, r2Result)
        
        # min of 相加結果
        r1Result = n1.r1 + n2.r1
        r2Result = n1.r2 + n2.r2
        self.rightSideLen = min(r1Result, r2Result)
        
    # 自行輸入已有結果   
    def setResult(self, sideName, leftSideLen, rightSideLen):
        self.sideName = sideName
        self.leftSideLen = leftSideLen
        self.rightSideLen = rightSideLen
        
    # 看內容
    def Inf(self):
        print("sidename : " + self.sideName)
        print("leftSideLen : " + str(self.leftSideLen))
        print("rightSideLen : " + str(self.rightSideLen))

# 以min of 相加結果對所有邊作排序        
def sortResult(results):
    for i in range(0,len(results)-1):
        for j in range(0,len(results)-1-i):
            if results[j].rightSideLen > results[j+1].rightSideLen:
                results[j], results[j+1] = results[j+1], results[j]

# 顯示結果
def printResults(results):
    print('\n#######################################')
    print('results')
    for i in range(0, len(results)):
        print(results[i].sideName, end = ' ')
        if i < len(results)-1:
            if results[i].rightSideLen < results[i+1].rightSideLen:
                # i的最大小於i+1的最小
                # 確定i邊小於i+1邊
                if results[i].rightSideLen < results[i+1].leftSideLen:
                    print('<', end = ' ')
                # 可能會相等但不會比i+1大
                else:
                    print('<=', end = ' ')
            # 無法準確判斷誰大誰小
            else:
                print(',', end = ' ')
    print('\n#######################################\n')
while True:
    # 做為目標的點
    targetNode = Node()
    print('Input targetNode Information')
    print('Name:')
    name = input()
    print('R1:')
    r1 = int(input())
    print('R2:')
    r2 = int(input())
    targetNode.setNode(name,r1,r2)

    # 其他點
    nodes = []
    tempNode = Node()
    print('Input other nodes: enter quit to quit')
    print('Name:')
    name = input()
    if name != 'quit':
        print('R1:')
        r1 = int(input())
        print('R2:')
        r2 = int(input())

    while name != 'quit':
        tempNode.setNode(name,r1,r2)
        nodes.append(tempNode)
        #tempNode.Inf()

        tempNode = Node()
        print('Input other nodes: enter quit to quit')
        print('Name:')
        name = input()
        if name == 'quit':
            break
        print('R1:')
        r1 = int(input())
        print('R2:')
        r2 = int(input())

    results = []
    for node in nodes:
        result = Result()
        result.setSideName_LeftLen_RightLen(targetNode, node)
        print('\n*********All results*********')
        result.Inf() #可以用來看每個邊的詳細結果
        results.append(result)
    
    # 題目給的QF = 40
    #result = Result()
    #result.setResult('QF', 40, 40)
    #results.append(result)
    
    sortResult(results)
    printResults(results)
    print('')
    print('Do you want to run another data? enter : Y/N')
    runornot = input()
    if runornot == 'n' or runornot == 'N':
        break
```
